## Global Meat Production in 2013

```python
df.head()
```

```python
df_global_meat = df[(df['Item']=='Meat')& (df['Element']=="Food")].copy()
```

```python
df_global_meat.head()
```

```python
index_list = df_global_meat.Y2013.sort_values(ascending=False).index
index_list
```

```python
production = []
for i in index_list:
    item = df_global_meat.Area.loc[i]
    production.append(item)
```

```python
plt.figure(figsize=(10,30))
ax = sns.barplot(x =df_global_meat.Y2013,y = df_global_meat.Area, data = df_global_meat, orient = 'h',
                order = production)
ax.set_title("Global Meat Production(Food) in 2013")
ax.set_ylabel('Country')
ax.set_xlabel('Production in 1000 tonnes')
plt.show()
```

## Global meat production per year

```python
years= list(df_global_meat.iloc[:,10:].columns)
year_list =[]
for i in years:
    #ano = i.replace("Y","")
    year_list.append(i)   
year_list[:5]
```

```python
len(year_list)
```

```python
country_list = list(df_global_meat.Area.unique())
country_list[:5]
```

```python
plt.figure(figsize=(15,15))
for country in country_list:
    yearly_production = []
    for year in year_list:
        yearly_production.append(df_global_meat[year][df_global_meat.Area==country].values[0])
    ax =sns.lineplot(x=year_list,y=yearly_production, label = country)

plt.legend(bbox_to_anchor=(0., 1.02, 1.5, .102), loc=3, ncol=8, mode="expand", borderaxespad=0.)
xtickNames = ax.set_xticklabels(year_list)
plt.setp(xtickNames, rotation=45,fontsize=10)
ax.set_title("Global Meat Production (Food)")
ax.set_xlabel('Year')
ax.set_ylabel('Production in 1000 tonnes')
plt.savefig('Global_meat.png')
plt.show()
```
